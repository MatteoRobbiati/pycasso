### `pycasso` tutorial

You can use your own `pycasso` artist to modify your images!
Let's start by importing your artistic workshop.

```python
from importlib import reload
from pycasso import workshop
```

```python
reload(workshop)
```

Where is my painter?!

Load it from the workshop!

While you do this, let the painter know about the picture you would like to bomb with your artistic impetus.

```python
picture = workshop.Painter("images/jolly.png")
```

Let's visualize it

```python
picture.image
```

OH no. 
This flag is too big for my shallow pirate vessel.

```python
picture.resize(size=(200,150))
picture.image
```

Uhm no, I liked it more before. Let's restore the original one.

```python
picture.back_to_original()
picture.image
```

I don't really like these colors, let's make it shiner.

```python
picture.replace_colors(
    old_color=(255, 255, 255),
    new_color=(255, 150, 255),
    alpha=255
)
```

```python
picture.image
```

Much better! 
Now, I would like to remove the black background, since I am planning to desing a logo for my new webpage! Since the background is black, I will remove the `(0, 0, 0)` parts.

```python
picture.remove_color(color=(0, 0, 0))
picture.image
```

Mmmm! This is weird. Why all the black is still there? Let's check the real nature of this evil color.

```python
picture.pixels[5,5]
```

Ahah! Found it. Some monster decided to use `(2, 3, 4)` instead of `(0, 0, 0)`.

```python
picture.remove_color(color=(2,3,4))
picture.image
```

Yes, this makes sense! But there are many shades of black. Let's do something better. We can set a shadow range $\mathcal{SR}:[c-\delta, c+\delta]$ around our color $c$ and remove all the pixels inside that range.

```python
picture.back_to_original()
picture.remove_color(color=(0,0,0), shadow_range=20)
```

```python
picture.image
```

Very good! But now, let's keep the black back on the bones!

```python
picture.replace_colors(
    old_color=(255, 150, 255),
    new_color=(0, 0, 0),
    alpha=255
)

picture.image
```

Save it!

```python
picture.save("images/cool_skull.png")
```
